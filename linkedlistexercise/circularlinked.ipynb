```python
class Node:
    def __init__(self,data,next=None):
        self.data = data
        self.next = next
```

```python
class CircularList:
    def __init__(self):
        self.head = None
        self.tail = None
        self.size = 0

    def isEmpty(self):
        if self.size == 0 and self.head is None and self.tail is None:
            return True
        else:
            return False
    
    def insert_at_head(self,data):
        newNode = Node(data)
        if self.head is None and self.tail is None:
            self.head = newNode
            self.tail = newNode
            self.head.next = self.tail
            self.size = self.size + 1
            return
        else:
            newNode.next = self.head
            self.tail.next = newNode
            self.head = newNode
            self.size = self.size + 1
    
    def insert_at_tail(self,data):
        newNode = Node(data)
        if self.isEmpty():
            self.head = newNode
            self.tail = newNode
            self.size = self.size + 1
            return
        else:
            newNode.next = self.head
            self.tail.next = newNode
            self.tail = newNode
            self.size = self.size + 1

    def delete_at_head(self):
        if self.isEmpty():
            print("List is empty")
            return
        elif self.head == self.tail:
            self.head = None
            self.tail = None
            self.size = 0
        else:
            temp = self.head.next
            self.tail.next = temp
            self.head = temp
            self.size = self.size - 1

    def delete_at_last(self):
        if self.isEmpty():
            print("List is empty")
            return
        elif self.tail == self.head:
            self.head = None
            self.tail = None
            self.size = 0
        else:
            temp = self.head

            while temp.next != self.tail:
                temp = temp.next

            temp.next = self.head
            self.tail = temp

    def listSize(self):
        print("List size: ",self.size,end="\n")

    def display(self):
        if self.isEmpty():
            print("List is empty")
            return
        else:
            print(self.head.data,end="-->")
            temp = self.head.next
            while temp != self.head:
                print(temp.data,end="-->")
                temp = temp.next

            print(temp.data,end="")
```

```python
cl = CircularList()

print(cl.isEmpty())

cl.insert_at_head(12)
cl.insert_at_head(13)
cl.insert_at_head(14)
cl.insert_at_head(15)
cl.insert_at_head(16)

cl.insert_at_tail(18)
cl.insert_at_tail(20)

cl.listSize()

cl.display()
print("\n")

cl.delete_at_head()
cl.delete_at_head()
# cl.delete_at_head()
# cl.delete_at_head()
# cl.delete_at_head()

cl.display()

print("\n")

cl.delete_at_last()
# cl.delete_at_last()
# cl.delete_at_last()

print("\n")

cl.listSize()

cl.display()
```
