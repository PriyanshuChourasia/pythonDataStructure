```python
# Node
class Node:
    def __init__(self,data,next=None):
        self.data = data
        self.next = next
```

```python
# Linked List
class LinkedList:
    def __init__(self):
        self.head = None
        self.size = 0

    def isEmpty(self):
        return self.size == 0 and self.head == None
    
    def length(self):
        return self.size
    
    def insertValues(self,data):
            newNode = Node(data)
            newNode.next = self.head
            self.head = newNode
            self.size += 1

    def insertAtPos(self,pos,data):
        if self.isEmpty():
            print("List is empty")
            return
        elif self.size < pos:
            print("Given position exceeds the length of list")
            return
        elif pos == 1:
            newNode = Node(data)
            newNode.next = self.head
            self.head = newNode
            self.size +=1
            return
        else:
            newNode = Node(data)
            temp = self.head
            i=1
            while temp.next != None:
                if i == pos - 1:
                    tempNext = temp.next
                    temp.next = newNode
                    newNode.next = tempNext
                    break
                temp = temp.next
                i = i + 1
            
            self.size +=1
        
    
    def deleteAtHead(self):
        if self.isEmpty():
            print("List is empty")
            return
        
        self.head = self.head.next
        self.size = self.size - 1

    def deleteAtPos(self,pos):
        if self.isEmpty():
            print("List is empty")
            return
        elif pos < 1:
            print("Please provide a valid position")
            return
        elif pos > self.size:
            print("Given position is not valid")
            return
        elif pos == 1:
            self.head = self.head.next
            self.size -= 1
            return
        else:
            temp = self.head
            i=1
            while temp:
                if i == pos - 1:
                    temp.next = temp.next.next
                temp = temp.next
                i = i + 1

            self.size -= 1

    def deleteAtEnd(self):
        if self.isEmpty():
            print("List is empty")
            return
        elif self.size == 1:
            self.head = None
            self.size = 0
            return
        else:
            temp = self.head

            while temp.next.next != None:
                temp = temp.next

            temp.next = None
            self.size = self.size - 1
            

    def insertAtHead(self,data):
        newNode = Node(data)
        newNode.next = self.head
        self.head = newNode
        self.size +=1

    def insertAtEnd(self,data):
        if self.isEmpty():
            print("List is empty")
            return

        temp = self.head
        newNode = Node(data)
        while temp.next != None:
            temp = temp.next
            
        temp.next = newNode

        self.size +=1
    
    def display(self):
        if self.isEmpty():
            print("List is empty")
            return
        
        print("\n")
        temp = self.head

        while temp:
            print(temp.data,end="-->")
            temp = temp.next

        print("None")
        
```

```python
# display list in reverser order

class ReverseList(LinkedList):
    
    def __init_subclass__(cls):
        return super().__init_subclass__()

    def reverseList(self):
        if self.isEmpty():
            print("List is empty")
            return
        
        temp = self.head
        newNode = None
        while temp:
            newNode = Node(temp.data,newNode)
            temp = temp.next
        
        self.head = newNode
    
```

```python
#  Reverse list

rl = ReverseList()

rl.insertValues(12)
rl.insertValues(123)
rl.insertValues(1234)
rl.insertValues(12345)

rl.display()

rl.reverseList()

print("\n")

rl.display()

print("\n")

print(rl.length())
```
