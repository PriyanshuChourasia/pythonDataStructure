```python
# Node
class Node:
    def __init__(self,data,next=None):
        self.data = data
        self.next = next
```

```python
# Linked List
class LinkedList:
    def __init__(self):
        self.head = None
        self.size = 0

    def isEmpty(self):
        return self.size == 0 and self.head == None
    
    def length(self):
        return self.size
    
    def insertValues(self,data):
        newNode = Node(data)
        if self.isEmpty():
            self.head = newNode
            self.size += 1
        else:
            newNode.next = self.head
            self.head = newNode
            self.size += 1

    def insertAtHead(self,data):
        newNode = Node(data)
        newNode.next = self.head
        self.head = newNode

    def insertAtEnd(self,data):
        if self.isEmpty():
            print("List is empty")
            return

        temp = self.head
        newNode = Node(data)
        while temp.next != None:
            temp = temp.next
            
        temp.next = newNode
    
    def display(self):
        if self.isEmpty():
            print("List is empty")
            return
        
        print("\n")
        temp = self.head

        while temp:
            print(temp.data,end="-->")
            temp = temp.next

        print("None")
        
```

```python
# create lists
ll = LinkedList()

print(ll.length())

print(ll.isEmpty())

ll.display()

ll.insertValues(23)
ll.insertValues(28)
ll.insertValues(39)

ll.display()

# insert at beginning
ll.insertAtHead(34)
ll.insertAtHead(57)
ll.insertAtHead(78)

ll.display()

# insert at end
ll.insertAtEnd(45)
ll.insertAtEnd(87)
ll.insertAtEnd(112)
ll.insertAtEnd(115)

ll.display()

```

```python
# display list in reverser order

class ReverseList(LinkedList):
    
    def __init_subclass__(cls):
        return super().__init_subclass__()

    def reverseList(self):
        if self.isEmpty():
            print("List is empty")
            return
        
        temp = self.head
        newNode = None
        while temp:
            newNode = Node(temp.data,newNode)
            temp = temp.next
        
        self.head = newNode
    
```

```python
#  Reverse list

rl = ReverseList()

rl.insertValues(12)
rl.insertValues(123)
rl.insertValues(1234)
rl.insertValues(12345)

rl.display()

rl.reverseList()

print("\n")

rl.display()

print("\n")

print(rl.length())
```
